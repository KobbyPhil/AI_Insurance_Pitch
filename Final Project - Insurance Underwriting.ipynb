### Feature Engineering
This is perhaps the most important part of this notebook.

Based on my industry knowledge, we know that these are high risk policies:

1. Old Age
2. Obese persons
3. High BMI
4. Extremely short or tall persons

We will therefore create few features such as:

1. Person very old or very young or in middle
2. Person very short or tall or in middle
3. Person with very high BMI or low BMI or in middle
4. Persons with obesity or are very thin or in middle

We will also create few more features such as:

1. Multiplication of BMI and Age - higher the factor, higher the risk
2. Multiplication of Weight and Age - higher the factor, higher the risk
3. Multiplication of Height and Age
4. Split of Product Info 2 into numericals and characters
5. Few stats description of Medical Keywords
6. Depending on the BMI, classification of lives into High Risk or Low Risk (very low and very high BMI both are risky factors)
7. Depending on BMI, Ht, Wt, Age, creation of more features which basically categorizes the risk into different buckets

```python
#1
df1['Product_Info_2_char'] = df1.Product_Info_2.str[0]
#2
df1['Product_Info_2_num'] = df1.Product_Info_2.str[1]

#3
df1['BMI_Age'] = df1['BMI'] * df1['Ins_Age']
#4
df1['Age_Wt'] = df1['Ins_Age'] * df1['Wt']
#5
df1['Age_Ht'] = df1['Ins_Age'] * df1['Ht']

med_keyword_columns = df1.columns[df1.columns.str.startswith('Medical_Keyword_')]
#6
df1['Med_Keywords_Count'] = df1[med_keyword_columns].sum(axis=1)

#7
df1['Ins_Age_sq'] = df1['Ins_Age'] * df1['Ins_Age']
#8
df1['Ht_sq'] = df1['Ht'] * df1['Ht']
#9
df1['Wt_sq'] = df1['Wt'] * df1['Wt']
#10
df1['BMI_sq'] = df1['BMI'] * df1['BMI']

#11
df1['Ins_Age_cu'] = df1['Ins_Age'] * df1['Ins_Age'] * df1['Ins_Age']
#12
df1['Ht_cu'] = df1['Ht'] * df1['Ht'] * df1['Ht']
#13
df1['Wt_cu'] = df1['Wt'] * df1['Wt'] * df1['Wt']
#14
df1['BMI_cu'] = df1['BMI'] * df1['BMI'] * df1['BMI']

# BMI Categorization
conditions = [
    (df1['BMI'] <= df1['BMI'].quantile(0.25)),
    (df1['BMI'] > df1['BMI'].quantile(0.25)) & (df1['BMI'] <= df1['BMI'].quantile(0.75)),
    (df1['BMI'] > df1['BMI'].quantile(0.75))]

choices = ['under_weight', 'average', 'overweight']
#15
df1['BMI_Wt'] = np.select(conditions, choices)

# Age Categorization
conditions = [
    (df1['Ins_Age'] <= df1['Ins_Age'].quantile(0.25)),
    (df1['Ins_Age'] > df1['Ins_Age'].quantile(0.25)) & (df1['Ins_Age'] <= df1['Ins_Age'].quantile(0.75)),
    (df1['Ins_Age'] > df1['Ins_Age'].quantile(0.75))]

choices = ['young', 'average', 'old']
#16
df1['Old_Young'] = np.select(conditions, choices)

# Height Categorization
conditions = [
    (df1['Ht'] <= df1['Ht'].quantile(0.25)),
    (df1['Ht'] > df1['Ht'].quantile(0.25)) & (df1['Ht'] <= df1['Ht'].quantile(0.75)),
    (df1['Ht'] > df1['Ht'].quantile(0.75))]

choices = ['short', 'average', 'tall']
#17
df1['Short_Tall'] = np.select(conditions, choices)

# Weight Categorization
conditions = [
    (df1['Wt'] <= df1['Wt'].quantile(0.25)),
    (df1['Wt'] > df1['Wt'].quantile(0.25)) & (df1['Wt'] <= df1['Wt'].quantile(0.75)),
    (df1['Wt'] > df1['Wt'].quantile(0.75))]

choices = ['thin', 'average', 'fat']
#18
df1['Thin_Fat'] = np.select(conditions, choices)

#19
df1['min'] = df1[med_keyword_columns].min(axis=1)
#20
df1['max'] = df1[med_keyword_columns].max(axis=1)
#21
df1['mean'] = df1[med_keyword_columns].mean(axis=1)
#22
df1['std'] = df1[med_keyword_columns].std(axis=1)
#23
df1['skew'] = df1[med_keyword_columns].skew(axis=1)
#24
df1['kurt'] = df1[med_keyword_columns].kurtosis(axis=1)
#25
df1['med'] = df1[med_keyword_columns].median(axis=1)
```

```python
# function to replace Missing Values
# Categorical Values are replaced with its model values whiles numeric values are replaced with its mean
from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):

    def __init__(self):
        """Impute missing values.

        Columns of dtype object are imputed with the most frequent value 
        in column.

        Columns of other types are imputed with mean of column.

        """
    def fit(self, X, y=None):

        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].mean() for c in X],
            index=X.columns)

        return self

    def transform(self, X, y=None):
        return X.fillna(self.fill)
```

```python
no_miss_train_data = DataFrameImputer().fit_transform(train_data)
```

```python
# Function to drop highly correlated features
def correlation(df, threshold):
    col_corr = set() # Set of all the names of deleted columns
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
                if colname in df.columns:
                    del df[colname] # deleting the column from the dataset

    print(df.shape)
```

```python
#drop features with correlation greater absolute value of 0.95
correlation(df_train, 0.95)
```

```python
#correctly format categorical feature and apply label encode
#Using domain knowledge, it could be seen that categorical features are of type - integer and object in the data
#Label encode categorical values

for y in df_train.columns:
    if (df_train[y].dtype == np.int64 or df_train[y].dtype == np.object):
        df_train[y] = df_train[y].astype('category', CategoricalDtype = df_train[y].unique()).cat.codes
```

### MODEL TRAINING, SELECTION AND ASSESSMENT

```python
#Split Data into train and test 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(df_train, Target, test_size = 0.2, random_state = 21, stratify = Target)
```

```python
#Pretraining Model with GBM
from sklearn.ensemble import GradientBoostingClassifier
gbrt = GradientBoostingClassifier(random_state=0)
gbrt.fit(X_train, y_train)
```

```python
#Check Accuracy of Model
print("Accuracy on training set: {:.3f}".format(gbrt.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt.score(X_test, y_test)))
```

```python
#Train with other Classifier Algiorithms
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
classifiers = [
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    KNeighborsClassifier(),
    LogisticRegression()
]
```

```python
classifier = []
score = []
testScore = []

for clf in classifiers:
    clf.fit(X_train, y_train)
    name = clf.__class__.__name__
    
    print("="*30)
    print(name)
    
    print('****Results****')
    
    acc = cross_val_score(clf, X_train, y_train, cv = 5).mean()
    print("Train Accuracy: {:.2%}".format(acc))
    
    test_score = clf.score(X_test, y_test)
    print("Test Accuracy: {:.2%}".format(test_score))
    
    classifier.append(name)
    score.append(acc)
    testScore.append(test_score)
```

```python
# create dataframe for model results
call = pd.Series(classifier)
ca = pd.DataFrame()
ca["Classifier"] = call

test = pd.Series(testScore)
ca["Test_Score"] = test

crs_val = pd.Series(score)
ca['Cross_Val_Score'] = crs_val
```

```python
#Sort values of train and test accurracy
ca.sort_values('Test_Score')
testResults = ca.sort_values('Test_Score').drop(['Cross_Val_Score'], axis = 1)
Cross_Val_Results = ca.sort_values('Cross_Val_Score').drop(['Test_Score'], axis = 1)
```

```python
#Plot Test Results
plt.figure(figsize=(12,5))
sns.barplot(x = 'Test_Score', y = 'Classifier', data = testResults)
```

```python
#Plot Train Results
plt.figure(figsize=(12,5))
sns.barplot(x = 'Cross_Val_Score', y = 'Classifier', data = Cross_Val_Results)
```

```python
# Gradient Boosting Classifier is selected
# finding best learning rate for the GBC
param_grid = {'learning_rate' : [0.001, 0.01, 0.1, 1]}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(GradientBoostingClassifier(), param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
# Best Parameter with best score
print("Best parameters: {}".format(grid_search.best_params_))
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))
```

```python
# Train model with best learning rate
gbm = GradientBoostingClassifier(learning_rate = 0.1)
gbm.fit(X_train, y_train)
```

```python
# Take a look at Important Features
importances = gbm.feature_importances_

data={'Feature_Name':df_train.columns,
      'Feature_Importance': importances
     }

feature_df=pd.DataFrame(data)

feature_df.sort_values(by=['Feature_Importance'],ascending=False,inplace=True)
best_features = feature_df.head(30)
fig, ax = plt.subplots(figsize=(20,25))
sns.barplot(data=best_features,y='Feature_Name',x='Feature_Importance')
```

### Model Assessment

```python
# Confusion Matrix
pred = gbm.predict(X_test)
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, pred)
print("Confusion matrix:\n{}".format(confusion))
```

```python
# Classification Report
from sklearn.metrics import classification_report
print(classification_report(y_test, pred, target_names = ['Rejected', 'Accepted']))
```

```python
# Precision and ROC Curve 
from sklearn.metrics import precision_recall_curve
precision, recall, thresholds = precision_recall_curve(y_test, gbm.decision_function(X_test))

close_zero = np.argmin(np.abs(thresholds))
plt.plot(precision[close_zero], recall[close_zero], 'o', markersize = 10, label = 'threshold zero', fillstyle = 'none',
        c =  'k', mew = 2)
plt.plot(precision, recall, label = 'precision recall curve')
plt.xlabel('Precision')
plt.ylabel('Recall')
```

```python
from sklearn.metrics import roc_curve 
fpr, tpr, thresholds = roc_curve(y_test, gbm.decision_function(X_test))
plt.plot(fpr, tpr, label="ROC Curve") 
plt.xlabel("FPR") 
plt.ylabel("TPR (recall)") # find threshold closest to zero 
close_zero = np.argmin(np.abs(thresholds)) 
plt.plot(fpr[close_zero], tpr[close_zero], 'o', markersize=10,         
label="threshold zero", fillstyle="none", c='k', mew=2) 
plt.legend(loc=4)

fig.savefig('ROC_Curve.png')
```
